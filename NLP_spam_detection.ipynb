```python
import pandas as pd
dataset=pd.read_csv("SMSSpamCollection",sep="\t",names=["labels","message"])
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
import nltk
import re
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(dataset)):
    msg=re.sub("[^a-zA-Z]"," ",dataset["message"][i])
    msg=msg.lower()
    msg=msg.split()
    msg=[ps.stem(word) for word in msg if not word in stopwords.words('english')]
    msg=' '.join(msg)
    corpus.append(msg)
```

```python
corpus[4]
```

```python
# creating bag of word model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
X=cv.fit_transform(corpus).toarray()
```

```python
len(X[0])
```

```python
X
```

```python
y=pd.get_dummies(dataset["labels"])
```

```python
y=y.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spammodel=MultinomialNB()
spammodel.fit(X_train,y_train)
```

```python
y_pred=spammodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
```

```python
score
```

```python
print(classification_report(y_test,y_pred))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
corpus = []
for i in range(0, len(dataset)):
    review = re.sub('[^a-zA-Z]', ' ', dataset['message'][i])
    review = review.lower()
    review = review.split()
    
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
lemmatizer.lemmatize("creeping")
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
```

```python
# training word2vec from scratch
model=gensim.models.Word2Vec(words,window=5,min_count=2)
```
